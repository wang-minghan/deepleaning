```python
# 多因子房价预测
import pandas as pd 
import numpy as np
data = pd.read_csv('kc_house_data.csv')
```

```python
data.head()
```

```python
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(20,20))
fig1 = plt.subplot(231)
plt.scatter(data.loc[:,"bedrooms"],data.loc[:,"price"])
plt.title("price vs bedrooms")
          
fig2 = plt.subplot(232)
plt.scatter(data.loc[:,"floors"],data.loc[:,"price"])
plt.title("price vs floors")

fig3 = plt.subplot(233)
plt.scatter(data.loc[:,"waterfront"],data.loc[:,"price"])
plt.title("price vs waterfront")

fig3 = plt.subplot(234)
plt.scatter(data.loc[:,"grade"],data.loc[:,"price"])
plt.title("price vs grade")

fig3 = plt.subplot(235)
plt.scatter(data.loc[:,"lat"],data.loc[:,"price"])
plt.title("price vs lat")
plt.show
```

```python
# define x&y
X =  data.loc[:,"bedrooms"]
y =  data.loc[:,"price"]
y.head()
```

```python
X = np.array(X).reshape(-1,1)
print(X.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR1 = LinearRegression()
LR1.fit(X,y)
```

```python
# price vs size
y_predict_1 = LR1.predict(X)
print(y_predict_1)
```

```python
# 迭代模型 evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

# 假设y和y_predict_1在您的代码的其他部分已经被定义：
mse1 = mean_squared_error(y, y_predict_1)
r2_score_1 = r2_score(y, y_predict_1)

print(mse1, r2_score_1)
```

```python
fig6 = plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X,y_predict_1,'r')
plt.show()
```
